```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
%matplotlib inline
np.random.seed(13)
```

6.1一様分布 1/(b-a) 
平均：(a+b)/2
標準偏差： (b-a)/sqrt(12)

```python
#6.1　np.random.uniformの場合
x1_1 = np.random.uniform(0.0, 10.0, 100000) # 0~10の乱数を100000個生成
plt.hist(x1_1, bins=50, normed=True)
print('mean: ', x1_1.mean(), 'std: ', x1_1.std())
plt.show();
```

```python
# np.random.randの場合
x1_2 = np.random.rand(100000) * 10
plt.hist(x1_2, 50, normed=True)
print('mean: ', x1_2.mean(), 'std: ', x1_2.std())
plt.show()
```

```python
#scipy.statsを使う場合
from scipy.stats import uniform
rv = uniform(loc = 0.0, scale=10.0) 
x1_3 = rv.rvs(size = 100000)
print('mean: ', x1_3.mean(), "std: ", x1_3.std())
plt.hist(x1_3, bins=50, normed=True)
plt.show();
```

ベルヌーイ分布 Θ, 1-Θ
平均：Θ
標準偏差：sqrt(Θ(1-Θ))

```python
#np.random.binominal(n=1)の場合
x2_1 = np.random.binomial(n=1, p=0.5, size=30)
plt.hist(x2_1, bins=3)
plt.show();
```

```python
#scipyの場合
from scipy.stats import bernoulli
x2_2 = bernoulli.rvs(0.5, size = 30)
plt.hist(x2_2, bins=3)
plt.show();
```

二項分布
平均：Nθ
標準偏差： sart(Nθ(1-θ))

```python
x3_1 = np.random.binomial(n = 30, p = 0.1, size=1000)
plt.hist(x3_1, bins=30, normed=True)
plt.show();
```

```python
from scipy.stats import binom
x3_1 = binom.rvs(n=30, p=0.5, size = 1000)
plt.hist(x3_1, bins=50, normed=True)
plt.show();
```
